### Use regex module to extract product ids from urls
#### product ids are used to send request to api
Product ids begin with p- then continues with a number of digits and ends with a ?.  
I compiled the pattern, then searched for it in each link in the dataframe.  
Used the return span to identify the index in the link  
Then I cleaned the returned p-2342894 like string by splitting it by "-", replacing the '?' at the end with nothing

```python
import re
```

```python
pattern = re.compile(r'p-\d+\?')
```

```python
product_ids = []
for item in df['Link'].index:
    matches = pattern.finditer(df['Link'][item])
    for i in matches:
        span = i.span()
        product_id = df['Link'][item][span[0]:span[1]].split('-')[1].replace('?','')
        product_ids.append(product_id)
    
```

```python
len(product_ids)
```

```python
df['product_id'] = product_ids
```
